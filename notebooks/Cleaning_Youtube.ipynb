### Bersihkan Teks Komentar

```python
def bersihkan_teks(teks):
    if pd.isna(teks):
        return ""
    teks = str(teks)

    teks = html.unescape(teks)                          # decode &amp; dll
    teks = teks.encode('ascii', 'ignore').decode('ascii') # hapus emoji & non-ASCII
    teks = re.sub(r'<br\s*/?>', ' ', teks)              # <br> → spasi
    teks = re.sub(r'<[^>]+>', '', teks)                 # hapus tag HTML
    teks = re.sub(r'http\S+|www\.\S+', '', teks)        # hapus URL
    teks = re.sub(r'@\w+', '', teks)                    # hapus mention
    teks = re.sub(r'#\w+', '', teks)                    # hapus hashtag
    teks = re.sub(r'\b\d+\b', '', teks)                 # hapus angka murni
    teks = re.sub(r'(.)\1{2,}', r'\1\1', teks)          # "bagussss" → "baguss"
    teks = re.sub(r'[!?.]{2,}', '.', teks)              # "!!!" → "."
    teks = re.sub(r'[^\w\s.,!?]', ' ', teks)            # hapus karakter spesial sisa
    teks = re.sub(r'\s+', ' ', teks).strip()            # rapikan spasi
    teks = teks.lower()                                  # huruf kecil semua

    return teks

df_clean = df_gabung.copy()
df_clean["Comment_Raw"] = df_clean["Comment"].copy()
df_clean["Comment_Clean"] = df_clean["Comment"].apply(bersihkan_teks)

df_clean = df_clean[
    df_clean["Comment_Clean"].str.strip() != ""
]

sebelum = len(df_clean)

df_clean = df_clean.drop_duplicates(
    subset=["Author","Comment_Clean","Time"],
    keep="first"
)

sesudah = len(df_clean)

print(f"✅ Cleaning selesai | Sebelum: {sebelum:,} | Dihapus: {sebelum-sesudah:,} | Sisa: {sesudah:,}")

for i, row in df_clean[["Comment_Raw","Comment_Clean"]].head(5).iterrows():
    print(
        f"{i+1}. "
        f"{row['Comment_Raw'][:70]} "
        f"→ {row['Comment_Clean'][:70]}"
    )
```

```python
from nlp_id.lemmatizer import Lemmatizer
import re

lemmatizer = Lemmatizer()

def normalisasi(teks):

    if pd.isna(teks):
        return ""

    teks = str(teks).lower()

    # angka kata
    teks = re.sub(
        r'(\w+)2\b',
        r'\1-\1',
        teks
    )

    # pakai kamus alay
    teks = " ".join(
        kamus_alay.get(kata, kata)
        for kata in teks.split()
    )

    # lemmatizer otomatis
    teks = lemmatizer.lemmatize(teks)

    # hapus kosong
    teks = " ".join(
        k for k in teks.split()
        if k.strip()
    )

    return teks
```

```python
df_clean["Comment_Norm"] = (
    df_clean["Comment_Clean"]
    .apply(normalisasi)
)

print("✅ Normalisasi selesai")
```

## Hapus Stopword

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords', quiet=True)

# Gabung stopword NLTK + tambahan manual bahasa Indonesia
stop_id = set(stopwords.words('indonesian'))
stop_tambahan = {
    'nya', 'yg', 'dg', 'dgn', 'sy', 'gw', 'gue', 'lo', 'lu',
    'bang', 'kak', 'mas', 'pak', 'bu', 'bro', 'sis', 'gan',
    'min', 'min', 'ka', 'bg', 'om', 'teh', 'sob',
    'nih', 'tuh', 'deh', 'dong', 'sih', 'lah', 'kan',
    'ah', 'oh', 'oo', 'oo', 'hm', 'eh'
}
stop_all = stop_id.union(stop_tambahan)

def hapus_stopword(tokens):
    return [kata for kata in tokens if kata not in stop_all]

df_clean['Tokens_Stop'] = df_clean['Tokens'].apply(hapus_stopword)

print("✅ Stopword dihapus")
print(f"   Contoh sebelum : {df_clean['Tokens'].iloc[2][:8]}")
print(f"   Contoh sesudah : {df_clean['Tokens_Stop'].iloc[2][:8]}")
```

## Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tokens):
    return [stemmer.stem(kata) for kata in tokens]

# ⚠️ Proses ini lambat karena 7.000+ baris — tunggu 2-5 menit
print("⏳ Stemming sedang berjalan, harap tunggu...")
df_clean['Tokens_Stem'] = df_clean['Tokens_Stop'].apply(stemming)

# Gabung kembali jadi string (untuk analisis sentimen lexicon)
df_clean['Comment_Final'] = df_clean['Tokens_Stem'].apply(lambda x: ' '.join(x))

print("✅ Stemming selesai!")
print(f"Contoh:")
for i, row in df_clean[['Comment_Clean','Comment_Final']].head(3).iterrows():
    print(f"  AWAL  : {row['Comment_Clean'][:80]}")
    print(f"  FINAL : {row['Comment_Final'][:80]}")
    print()
```
