## STANDARISASI KOLOM DAN MERGE DATASET

```python
# SHOPEE
df_shopee = pd.DataFrame()
df_shopee['brand'] = shopee['judul_produk']
df_shopee['price'] = shopee['harga']
df_shopee['lokasi'] = shopee['kota']
df_shopee['rating'] = shopee['rating']
df_shopee['terjual'] = shopee['terjual']
df_shopee['ecommerce'] = 'Shopee'

# TOKOPEDIA
df_tokopedia = pd.DataFrame()
df_tokopedia['brand'] = tokopedia['type_hp']
df_tokopedia['price'] = tokopedia['harga']
df_tokopedia['lokasi'] = tokopedia['lokasi_toko']
df_tokopedia['rating'] = tokopedia['rating']
df_tokopedia['terjual'] = tokopedia['terjual']
df_tokopedia['ecommerce'] = 'Tokopedia'

# LAZADA
df_lazada = pd.DataFrame()
df_lazada['brand'] = lazada['brand']
df_lazada['price'] = lazada['harga']
df_lazada['lokasi'] = lazada['lokasi_toko']
df_lazada['rating'] = lazada['jumlah_rating']
df_lazada['terjual'] = lazada['penjualan']
df_lazada['ecommerce'] = 'Lazada'
```

```python
ecommers = pd.concat(
    [df_shopee, df_tokopedia, df_lazada],
    ignore_index=True
)
```

## CLEANING

#### Cleaning Kolom Brand

```python
ecommers['brand'] = (
    ecommers['brand']
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
)

# jika ada brand kosong atau tulisan nan hasil konversi string
ecommers['brand'] = ecommers['brand'].replace('nan', pd.NA)
```

#### Cleaning Kolom Price

```python
ecommers['price'] = (
    ecommers['price']
    .astype(str)
    .str.replace('Rp', '', regex=False)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip()
)

# mengubah ke numerik, jika gagal akan menjadi NaN
ecommers['price'] = pd.to_numeric(ecommers['price'], errors='coerce')
```

#### Cleaning Kolom Rating

```python
ecommers['rating'] = (
    ecommers['rating']
    .astype(str)
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.strip()
)

# mengubah ke numerik, jika gagal akan menjadi NaN
ecommers['rating'] = pd.to_numeric(ecommers['rating'], errors='coerce')
```

#### Cleaning Kolom Terjual

```python
ecommers['terjual'] = ecommers['terjual'].astype(str).str.lower()
ecommers['terjual'] = ecommers['terjual'].str.replace('terjual', '', regex=False)
ecommers['terjual'] = ecommers['terjual'].str.replace('sold', '', regex=False)
ecommers['terjual'] = ecommers['terjual'].str.replace('+', '', regex=False)
```

#### Konversi Format Khusus Terjual

```python
def convert_terjual_to_numeric(value):
    value = str(value).strip()
    if 'rb' in value:
        try:
            return float(value.replace('rb', '').strip()) * 1000
        except ValueError:
            return None
    elif 'k' in value:
        try:
            return float(value.replace('k', '').strip()) * 1000
        except ValueError:
            return None
    else:
        try:
            return float(value.replace(',', ''))
        except ValueError:
            return None
```

```python
ecommers['terjual'] = ecommers['terjual'].apply(convert_terjual_to_numeric)

# mengubah ke numerik, jika gagal akan menjadi NaN
ecommers['terjual'] = pd.to_numeric(ecommers['terjual'], errors='coerce')
```

## PENANGANAN MISSING VALUE

```python
# Price -> median
median_price = ecommers['price'].median()
ecommers['price'] = ecommers['price'].fillna(median_price)
```

```python
# Lokasi -> mode
mode_lokasi = ecommers['lokasi'].mode()[0]
ecommers['lokasi'] = ecommers['lokasi'].fillna(mode_lokasi)
```

```python
# Rating -> median
median_rating = ecommers['rating'].median()
ecommers['rating'] = ecommers['rating'].fillna(median_rating)
```

```python
# Terjual -> imputasi nilai 0
ecommers['terjual'] = ecommers['terjual'].fillna(0)
```

```python
# Brand -> isi dengan 'Tidak diketahui'
ecommers['brand'] = ecommers['brand'].fillna('Tidak diketahui')
```

## ANALISIS OUTLIER

```python
Q1_price = ecommers['price'].quantile(0.25)
Q3_price = ecommers['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

batas_bawah_price = Q1_price - 1.5 * IQR_price
batas_atas_price = Q3_price + 1.5 * IQR_price

outlier_price = ecommers[(ecommers['price'] < batas_bawah_price) | (ecommers['price'] > batas_atas_price)]
print("Jumlah outlier price:", outlier_price.shape[0])

Q1_rating = ecommers['rating'].quantile(0.25)
Q3_rating = ecommers['rating'].quantile(0.75)
IQR_rating = Q3_rating - Q1_rating

batas_bawah_rating = Q1_rating - 1.5 * IQR_rating
batas_atas_rating = Q3_rating + 1.5 * IQR_rating

outlier_rating = ecommers[(ecommers['rating'] < batas_bawah_rating) | (ecommers['rating'] > batas_atas_rating)]
print("Jumlah outlier rating:", outlier_rating.shape[0])

Q1_terjual = ecommers['terjual'].quantile(0.25)
Q3_terjual = ecommers['terjual'].quantile(0.75)
IQR_terjual = Q3_terjual - Q1_terjual

batas_bawah_terjual = Q1_terjual - 1.5 * IQR_terjual
batas_atas_terjual = Q3_terjual + 1.5 * IQR_terjual

outlier_terjual = ecommers[(ecommers['terjual'] < batas_bawah_terjual) | (ecommers['terjual'] > batas_atas_terjual)]
print("Jumlah outlier terjual:", outlier_terjual.shape[0])
```
